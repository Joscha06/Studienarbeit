### Im folgenden werde ich eine erste Implementierung eines Gradient Boosting Trees auf unserem Datensatz laufen lassen

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd

# Schritt 1: Daten laden(easy)
train_values = pd.read_csv('train_values.csv')
test_values = pd.read_csv('test_values.csv')
train_labels = pd.read_csv('train_labels.csv')

# Schritt 2: Trainingsdaten und Labels zusammenführen(logisch)
train_data = pd.merge(train_values, train_labels, on='building_id')

# Schritt 3: Aufteilung in Features (X) und Zielvariable (y)
X_train = train_data.drop(['building_id', 'damage_grade'], axis=1)
y_train = train_data['damage_grade']

# Schritt 4: Gradient Boosting Tree-Modell erstellen und anpassen (keine Ahnung ob die Werte so passen)
# funktioniert wegen den kategorischen Attributen noch nicht, label encoding? one hot encoding?(bisher angwendet, nicht so sinnvoll)
n_estimators = 100
learning_rate = 0.1
max_depth = 3

gbt_model = GradientBoostingRegressor(n_estimators=n_estimators, learning_rate=learning_rate, max_depth=max_depth, random_state=42)
gbt_model.fit(X_train, y_train)

# Schritt 5: Testdaten vorbereiten und Vorhersagen machen
X_test = test_values.drop('building_id', axis=1)
y_pred = gbt_model.predict(X_test)

```
